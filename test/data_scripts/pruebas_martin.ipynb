# PRUEBAS MARTIN SALIBE 

```python
import matplotlib.pylab as plt
import numpy as np 




list_files = [
    '../../../results/prueba_martin/TC-MS-SR-0.80.txt',
    '../../../results/prueba_martin/TC-MS-SR-1.20.txt',
    #'../../../results/prueba_martin/TC-MS-SR-2.00.txt',
    '../../../results/prueba_martin/TC-MS-R1M-0.5.txt',
    '../../../results/prueba_martin/TC-MS-R1M-1.20.txt',
    '../../../results/prueba_martin/TC-MS-R1M-2.00.txt'
]

raw_adc_rp2040_hex = np.zeros((2000,len(list_files)),dtype=int)
raw_adc_rp2040=np.zeros((2000,len(list_files)),dtype=int)
n_samples_adc = np.arange(0,2000,1.0,dtype=float) # eje x para gráficas ADC 
mean_raw_adc = [] 
max_raw_adc = [] 
min_raw_adc = [] 


i = 0 
for file in list_files: 
    d0,d1,d2 = np.genfromtxt(file, delimiter=',', dtype=None, encoding=None,unpack=True)
    d1 =[int(str(x),16) for x in d1]
    #for x in d1:
    #    print(int(x,16)) 
    d2 =d2.astype(int) 
    raw_adc_rp2040[:,i] = d2 
    raw_adc_rp2040_hex[:,i] = d1 
    mean_raw_adc.append(np.mean(d2))
    max_raw_adc.append(np.amax(d2))
    min_raw_adc.append(np.amin(d2))
    i = i +1 


#### plotting file #### 
titles = ['Voltage = 0.8','Voltage = 1.2', 'Voltage = 2.0']
figure080,axes080 = plt.subplots(figsize=(20,10))
figure080.suptitle(titles[0],fontsize = 40)
axes080.plot(n_samples_adc,raw_adc_rp2040[:,0],label = 'SR')
axes080.plot(n_samples_adc,raw_adc_rp2040[:,2],label = r'R100M$\Omega$')
axes080.legend(fontsize=30) 


figure120,axes120 = plt.subplots(figsize=(20,10))
figure120.suptitle(titles[1],fontsize = 40)
axes120.plot(n_samples_adc,raw_adc_rp2040[:,1],label = 'SR')
axes120.plot(n_samples_adc,raw_adc_rp2040[:,3],label = r'R100$\Omega$')
axes120.legend(fontsize=30) 




```
